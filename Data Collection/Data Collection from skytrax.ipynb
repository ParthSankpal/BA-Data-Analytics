```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

def scrape_reviews(url):
    reviews = []
    stars = []
    date = []
    country = []

    for i in range(1, 36):
        page = requests.get(f"{url}/page/{i}/?sortby=post_date%3ADesc&pagesize=100")
        soup = BeautifulSoup(page.content, "html.parser")

        reviews += [item.text for item in soup.find_all("div", class_="text_content")]

        for item in soup.find_all("div", class_="rating-10"):
            stars.append(item.span.text if item.span else "None")

        date += [item.text for item in soup.find_all("time")]
        country += [item.span.next_sibling.text.strip(" ()") for item in soup.find_all("h3")]

    return reviews, stars, date, country

# Call the function to scrape data
url = "https://www.airlinequality.com/airline-reviews/british-airways"
reviews, stars, date, country = scrape_reviews(url)

# Ensure the lengths match
min_length = min(len(reviews), len(stars), len(date), len(country))
reviews, stars, date, country = reviews[:min_length], stars[:min_length], date[:min_length], country[:min_length]

# Create a DataFrame
df = pd.DataFrame({"reviews": reviews, "stars": stars, "date": date, "country": country})

# Display a preview of the DataFrame
df
```
